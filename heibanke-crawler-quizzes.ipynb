## Lesson 1 (parsing html)

```python
# regular expression practice
import re

s = '<h3>下一个你需要输入的数字是83105. </h3>'

pattern = re.compile(r'<h3>\D+(\d{5})\D*</h3>')
result = pattern.findall(s)

print(result)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import re

def looper(num=''):
    try:
        url = 'http://www.heibanke.com/lesson/crawler_ex00/%s'%num
        print('processing: ' + url)
        
        r = requests.get(url)
        html = r.text
        
        ptn = re.compile(r'<h3>\D+(\d{5})\D*</h3>')
        m = ptn.findall(html)
        print(m)

        if len(m) > 0:
            looper(m[0])
        else:
            print('Didn"t match any number.')
            print(html)
            
    except Exception as e:
        print(e)
        return

looper()
```

## Lesson 2 (Brute force)

```python
import requests
import re

def brute_force(num=0):
    print(num)
    if num > 30: return
    
    url = 'http://www.heibanke.com/lesson/crawler_ex01/'
    
    payload = {'password': num, 'username':'abc', 'csrfmiddlewaretoken':'G702vtIfMDT5IOGf9doOGK82aB0pmgK6'}
    r = requests.post(url, data=payload)
    
    ptn = re.compile(r'<h3>.*错误')
    m = ptn.findall(r.text)
    print(m)
    
    if len(m) > 0:
        brute_force(num+1)
    else:
        print(r.text)


brute_force()
```
